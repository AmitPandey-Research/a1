##Q-8 : Gamma Transformation/Power law transformation

```
# importing all the neccessary libraries. dont forget to mount the drive

import numpy as np
import cv2 # to read files / images using cv2
import os #importing operating system 
import matplotlib.pyplot as plt
import matplotlib.image as img
from google.colab.patches import cv2_imshow # in new version cv2.imshow doesnt work
IMAGES_FOLDER_PATH = "/content/drive/My Drive/images" # creating a shorter name to store the path
```

```
# reading all the images :
stealth = cv2.imread(os.path.join(IMAGES_FOLDER_PATH,"stealth.png"))
unstealth= cv2.imread(os.path.join(IMAGES_FOLDER_PATH,"unstealth.png"))
aston= cv2.imread(os.path.join(IMAGES_FOLDER_PATH,"aston_invisible.jpg"))
```

```
#converting images to grayscale
stealth_gray = cv2.cvtColor(stealth, cv2.COLOR_BGR2GRAY)
unstealth_gray = cv2.cvtColor(unstealth, cv2.COLOR_BGR2GRAY)
aston_gray = cv2.cvtColor(aston, cv2.COLOR_BGR2GRAY)
```

```
plt.imshow(stealth, cmap='gray',vmin=0,vmax=255)
```

```
def gammaTranform(img, gamma):
  ''' gamma > 1 to darken and <1 to brighten image'''
  imgnew = 255*((img/255)**gamma)
  imgfinal = np.ceil(imgnew)
  return imgfinal
```

```
# gamma <1 for brightening the image
transformed_stealth = gammaTranform(stealth_gray,0.5)
```

```
plt.imshow(transformed_stealth, cmap='gray',vmin=0,vmax=255)
```

```
# gamma >1 for darkening the image
transformed_unstealth = gammaTranform(unstealth_gray,2.5)
```

```
plt.imshow(transformed_unstealth, cmap='gray',vmin=0,vmax=255)
```

```
# gamma <1 for brightening the image
transformed_am = gammaTranform(aston_gray,0.3)
```

```
plt.imshow(transformed_am, cmap='gray',vmin=0,vmax=255)
```
