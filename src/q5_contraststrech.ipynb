```
# importing all the neccessary libraries. dont forget to mount the drive

import numpy as np
import cv2 # to read files / images using cv2
import os #importing operating system 
import matplotlib.pyplot as plt
import matplotlib.image as img
from google.colab.patches import cv2_imshow # in new version cv2.imshow doesnt work
IMAGES_FOLDER_PATH = "/content/drive/My Drive/images" # creating a shorter name to store the path
```

```
qb= cv2.imread(os.path.join(IMAGES_FOLDER_PATH,"quantum_bad.jpg"))
```

```
max_intensity = np.max(qb, axis=(0,1))
min_intensity = np.min(qb, axis=(0,1))
```

```
def linContrastStretching (Im,max,min):
  copyqb = np.copy(Im)
  channel = 3
  for i in range(channel) :
    copyqb[:,:, i] = min + (max-min)*(( copyqb[:,:, i] ) / ( max_intensity[i])-min_intensity[i] )
  return copyqb
```

```
qb_improved = linContrastStretching(qb,255,0)
```

```
rgbqb = cv2.cvtColor(qb_improved, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(rgbqb)
```

## Linear contrast stretching:
### Linear contrast stretching is performed to improve overall contrast of the image by mapping the original intensity values(having a smaller range,say 50-180 on grayscale) over a greater range of values (0 to 255).
### this is done by using = 
### $ f(a) = a_{min} + (a - a_{low})*[(a_{max} - a_{min})/(a_{high}-a_{low})]$
#### $a_{max} = maximum intensity in original image,$
####$ a_{min} = minimum intensity in original image,$
####$(a_{high}= maximum intensity in mapped image,$
####$ a_{low}= minimum intensity in mapped image$
